After you've trained a neural network to detect facial keypoints, you can then apply this network to *any* image that includes faces. The neural network expects a Tensor of a certain size as input and, so, to detect any face, you'll first have to do some pre-processing.

1. Detect all the faces in an image using a face detector (we'll be using a Haar Cascade detector in this notebook).
2. Pre-process those face images so that they are grayscale, and transformed to a Tensor of the input size that your net expects. This step will be similar to the `data_transform` you created and applied in Notebook 2, whose job was tp rescale, normalize, and turn any iimage into a Tensor to be accepted as input to your CNN.
3. Use your trained model to detect facial keypoints on the image.

#### Loop over the faces found by Haar, process, infer CNN to predict keypoints, save ROI and keypoints

```python
image_copy = np.copy(image)

# the transforms we defined in Notebook 1 are in the helper file `data_load.py`
from data_load import Rescale, Normalize, ToTensor

normalize = Normalize()
rescale = Rescale(224)
toTensor = ToTensor()
margin = 30

def pre_process_image(x, y, w, h, image):
    # Select the region of interest that is the face in the image 
    roi = image[y-margin : y+h+margin, x-margin : x+w+margin]
    
    # Rescale the detected face to be the expected square size for your CNN (224x224, suggested)
    roi = rescale({"image": roi, "keypoints": 0})["image"]
    
    # Normalize the grayscale image so that its color range falls in [0,1] instead of [0,255]
    roi = normalize({"image": roi, "keypoints": 0})["image"]
                
    # Reshape the numpy image shape (H x W x C) into a torch image shape (C x H x W)
    roi = toTensor({"image": roi, "keypoints": np.array([0])})["image"]
    return roi.unsqueeze(0)

def predict_keypoints(roi):
    # convert images to FloatTensors
    roi_tensor = roi.type(torch.FloatTensor)
    output_pts = net(roi_tensor)
    # reshape to batch_size x 68 x 2 pts
    output_pts = output_pts.view(output_pts.size()[0], 68, -1)

    output_pts = output_pts.detach().numpy()
    # undo normalization of keypoints  
    output_pts = output_pts * 50.0 + 100
    
    output_pts = np.squeeze(output_pts)
    return output_pts

def torn_to_image(roi):
    roi = np.squeeze(roi, 1)
    roi = np.transpose(roi, (1, 2, 0))
    return np.squeeze(roi)


roi_list = []
output_pts_list = []

# loop over the detected faces from your haar cascade
for i, (x,y,w,h) in enumerate(faces):
    # Pre process image
    roi = pre_process_image(x, y , w , h, image_copy)
    
    # Make facial keypoint predictions using your loaded, trained network 
    # perform a forward pass to get the predicted facial keypoints        
    keypoints = predict_keypoints(roi)
    
    # convert torn image to rgb pattern
    roi_copy = torn_to_image(np.copy(roi))

    # save roi
    roi_list.append(roi_copy)
    
    # save output points 
    output_pts_list.append((x, y, keypoints))
```

#### Display Facial Keypoint inside of the original image

```python
# define plot settings and display the image where it will drawn the keypoints of each face
fig = plt.figure(figsize=(10, 8))
plt.imshow(image_copy, cmap='gray')

for i in range(len(roi_list)):
    (x, y, keypoints) = output_pts_list[i]
    roi = roi_list[i]
    
    x = keypoints[:, 0] + x - margin
    y = keypoints[:, 1] + y - margin

    # Display each detected face and the corresponding keypoints
    plt.scatter(x, y, s=50, marker='.', c='b'), plt.xticks([]), plt.yticks([])

plt.show()
```

#### Display Facial Keypoint in each ROI (Region Of Interest)

```python
for i in range(len(roi_list)):
    (x, y, keypoints) = output_pts_list[i]
    roi = roi_list[i]
    
    x = keypoints[:, 0]
    y = keypoints[:, 1]
    
    # Display each detected face and the corresponding keypoints
    plt.subplot(100+(len(roi_list))*10+1+i)
    plt.imshow(roi, cmap='gray'), plt.xticks([]), plt.yticks([])
    plt.scatter(x, y, s=50, marker='.', c='b')

plt.show()
```
